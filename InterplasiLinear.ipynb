```python
import numpy as np
import matplotlib.pyplot as plt

# Meminta pengguna untuk memasukkan nilai x dan y
x_str = input('Masukkan nilai x, dipisahkan dengan spasi: ')
y_str = input('Masukkan nilai y, dipisahkan dengan spasi: ')

# Mengubah string input menjadi list nilai x dan y
x = [float(val) for val in x_str.split()]
y = [float(val) for val in y_str.split()]

# Menghitung titik-titik pada kurva interpolasi linier
x_interp = np.linspace(x[0], x[-1], 100)
y_interp = np.interp(x_interp, x, y)

# Meminta pengguna untuk memasukkan nilai x yang ingin diprediksi
x_pred_str = input('Masukkan nilai x yang ingin diprediksi: ')
x_pred = float(x_pred_str)

# Menghitung nilai y yang diprediksi menggunakan interpolasi linier
y_pred = np.interp(x_pred, x, y)
print("Nilai X : ",x)
print("Nilai Y : ",y)
print("Nilai Prediksi X: ", x_pred_str)
print(f"Hasil Nilai Prediksi adalah {y_pred}")

# Menampilkan plot titik-titik data dan kurva interpolasi linier
plt.plot(x, y, 'o', label='Data')
plt.plot(x_interp, y_interp, label='Interpolasi')
plt.plot(x_pred, y_pred, 'ro', label=f'Prediksi ({x_pred}, {y_pred:.2f})')
plt.title('Kurva Interpolasi Linier')
plt.xlabel('Nilai X')
plt.ylabel('Nilai Y')
plt.legend()
plt.show()
```
